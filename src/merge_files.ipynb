```python
import os
import glob
import pandas as pd

def get_merged_csv(flist, **kwargs):
    try:
        return pd.concat([pd.read_csv(f, **kwargs) for f in flist], axis=1, ignore_index=True)
    except:
        print(flist)
        pass

# p = ['경기도 가평군 가평읍', '경기도 가평군 북면']

for k in lname:
    df = get_merged_csv(glob.glob('./data/climate_data/'+k+'/*.csv'), index_col=None)
    try:
        df.to_excel('data/addr/'+k+'.xlsx', encoding='utf-8', index=False)
    except:
        print(k)
        pass

print(df.head())
```
