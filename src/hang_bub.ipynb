### 법정동을 행정동으로 변환하는 주소 전처리 작업

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("data/주소리스트.xlsx")
display(data.head(10))
```

```python
data = data.drop(data.columns[[0, 4, 6, 7]], axis=1) # 행정동코드, 법정동코드, 생성일자, 말소일자 제거
data.drop([0, 1], inplace=True) # 필요없는 행 제거
```

```python
display(data.head())
```

```python
data.set_index('시도명', inplace=True)
data.reset_index(inplace=True) # 행 번호 초기화(0부터 시작)
```

```python
display(data.head()) # 인덱스번호가 리셋되었음
```

```python
data['행정동주소'] = data['시도명'] + " " + data['시군구명'] + " " + data['읍면동명']
data['법정동주소'] = data['시도명'] + " " + data['시군구명'] + " " + data['동리명']
display(data.head())
```

```python
data1 = data.drop(data.columns[[0, 1, 2, 3]], axis=1) # 행정동주소와 법정동주소 남기고 다 제거
```

```python
display(data1.head())
```

#### 주소를 id로 잡고 merge 하기 전에 잘못된 데이터 수정하기 

```python
dong.replace("영등포동", "영동포동") # 영등포동 --> 영동포동으로 
```

```python
dong.columns=['법정동주소', '재산피해소계'] 
```

```python
hb = pd.merge(dong, data, on='법정동주소', how='left')
display(hb.head())
```

```python
len(hb)
```

```python
hb[hb['행정동주소'].isnull()]
```

```python
hb.to_csv('data/hb.csv', encoding='utf-8')
```

#### null값 직접 전처리 후 import

```python
dong = pd.read_csv("data/new_hb.csv", encoding='utf-8')
dong = dong.drop(dong.columns[[0]], axis=1)
display(dong.head(10))
```

```python
dong = dong.drop(dong.columns[[0]], axis=1) #이제 필요없는 법정동 제거
```

```python
dong[dong['재산피해소계'].isnull()] # null값 있는지 확인 --> 없음. 데이터 매치 끝.
```

```python
len(dong)
```

```python
display(dong.head())
```

```python
dong.duplicated(['행정동주소']) # 행정동으로 주소를 변환하니 중복값이 생겼다.
```

```python
all_dong = dong.groupby(['행정동주소']).sum('재산피해소계')
print(all_dong.head())
```

```python
len(all_dong)
```

```python
all_dong.drop_duplicates(['재산피해소계'], keep='first', inplace=True)
```

```python
len(all_dong)
```

```python
all_dong.to_csv('data/all_dong.csv', encoding='utf-8')
```
