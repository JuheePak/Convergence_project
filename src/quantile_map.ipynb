```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/소방청_연간화재통계_20191231.csv", encoding='ansi')
display(data.head(5))
```

```python
data_1 = data[data['장소대분류']=='산업시설']
data_1.head()
```

```python
len(data_1)
```

```python
data2 = data_1.loc[:, ['시도', '시·군·구', '인명피해(명)소계', '읍면동', '재산피해소계']]
data2.head()
```

```python
data2['주소'] = data2['시도'] + " " + data2['시·군·구'] + " " + data2['읍면동']
display(data2.head())
```

```python
data3 = data2.loc[:, ['주소', '재산피해소계', '인명피해(명)소계']]
data3.head()
```

```python
all = data3.groupby(['주소']).sum(['재산피해소계', '인명피해(명)소계'])
print(all.head())
```

```python
sort_재산 = all.sort_values(by=['재산피해소계'], ascending=False).head()
display(sort_재산.head())
```

```python
sort_인명 = all.sort_values(by=['인명피해(명)소계'], ascending=False).head()
display(sort_인명.head())
```

```python
all['quantile_재산'] = pd.qcut(all['재산피해소계'], q=4, labels=['재산피해_하위1', '재산피해_하위2', '재산피해_하위3', '재산피해_하위4'])
all.sort_values(by='재산피해소계')
```

```python
import pandas as pd
import jenkspy
```

```python
all['cut_jenks_인명'] = pd.cut(
    all['인명피해(명)소계'],
    bins=jenkspy.jenks_breaks(all['인명피해(명)소계'], nb_class=2),
    labels=['인명피해_하위1', '인명피해_하위2'])

all.sort_values(by='인명피해(명)소계')
all.to_excel('./data/group_data.xlsx')
```

```python
all.groupby(['quantile_재산', 'cut_jenks_인명']).count()
```
